```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd


result = pd.read_csv('../data/BBL.csv')

fig, ax = plt.subplots()

x = result.index
line, = ax.plot(x, result['stock_price'])


# result['stock_price'].plot(figsize=(10,6), c='grey', lw='2', alpha=1);
# result['marker_buy'].plot(style='o', ms=5, label='buy', c='g', alpha=0.5);
# result['marker_sell'].plot(style='o', ms=5, label='sell', c='r', alpha=0.5);


def animate(i):
    line.set_ydata(result['stock_price'].iloc[i])  # update the data.
    return line,


ani = animation.FuncAnimation(fig, animate, interval=20, blit=True, save_count=50)

# To save the animation, use e.g.
#
# ani.save("movie.mp4")
#
# or
#
# writer = animation.FFMpegWriter(
#     fps=15, metadata=dict(artist='Me'), bitrate=1800)
# ani.save("movie.mp4", writer=writer)

plt.show()
```
