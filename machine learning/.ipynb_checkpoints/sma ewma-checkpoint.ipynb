```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pandas_datareader as web
```

```python
df = web.DataReader('aapl', 'yahoo', '2018-01-01', '2021-01-01')
```

```python
df['10-day-SMA'] = df['Adj Close'].rolling(10).mean()
df['20-day-SMA'] = df['Adj Close'].rolling(20).mean()
df['50-day-SMA'] = df['Adj Close'].rolling(50).mean()
df['100-day-SMA'] = df['Adj Close'].rolling(100).mean()
```

```python
df[['Adj Close', '10-day-SMA', '20-day-SMA', '50-day-SMA', '100-day-SMA']].iloc[-200:].plot()
```

```python
df['10-day-EWA'] = df['Adj Close'].ewm(span=10).mean()
df['20-day-EWA'] = df['Adj Close'].ewm(span=20).mean()
df['50-day-EWA'] = df['Adj Close'].ewm(span=50).mean()
df['100-day-EWA'] = df['Adj Close'].ewm(span=100).mean()
```

```python
df[['Adj Close', '10-day-EWA', '20-day-EWA', '50-day-EWA', '100-day-EWA']].iloc[-200:].plot()
```
