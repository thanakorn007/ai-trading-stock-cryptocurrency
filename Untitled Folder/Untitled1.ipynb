```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
stock_name = 'bbl'.upper()
```

```python
temp_dir = {}
stat = []

url = 'https://finance.yahoo.com/quote/'+ stock_name + '/key-statistics?p=' + stock_name
page = requests.get(url)
page_content = page.content
soup = BeautifulSoup(page_content, 'html.parser')
tabl = soup.find_all('div', {"class" : "Mstart(a) Mend(a)"})
for t in tabl:
    rows = t.find_all('div')
    for row in rows:
        data = row.get_text(separator='|').split('|')
        stat.append(data)
        temp_dir[data[0]] = [text for text in data[1:] if len(text) > 1]


for table in temp_dir: 
    for i in range(len(temp_dir[table])):
        if temp_dir[table][i][0] == '(':
            temp_dir[table][i-1] = temp_dir[table][i-1] + ' '+ temp_dir[table][i]
            
    temp_dir[table] = [text for text in temp_dir[table] if text[0] != '(']
    
del temp_dir['Market Cap (intraday)']
del temp_dir['Financial Highlights']
del temp_dir['Trading Information']
del temp_dir['']

for table in temp_dir:
    key = []
    value = []
    for i in range(0, len(temp_dir[table]), 2):
        key.append(temp_dir[table][i])
        value.append(temp_dir[table][i+1])

    temp_dir[table] = pd.DataFrame(data=value, index=key, columns=['value'])
    
for key in temp_dir:
    print(key)
    print(temp_dir[key])
    print('==================================================================')
```
